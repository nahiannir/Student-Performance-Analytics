```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Students.csv') # load csv file into dataframe
df.head() # shows first 5 rows
```

```python
df.info() # shows column names, data types and non null counts
df.describe() # show statistical summary (mean, min, max , etc,.)
```

```python
# calculate average marks
df['Average'] = df[['Math','Physics','Chemistry','English']].mean(axis=1)
# Calculate total marks
df['Total'] = df[['Math','Physics','Chemistry','English']].sum(axis=1)
```

```python
def grade(avg):  # function to assign grade based on average
    if avg >= 80:
        return 'A+'
    elif avg >= 70:
        return 'A'
    elif avg >= 60:
        return 'B'
    elif avg >= 50:
        return 'C'
    else:
        return 'F'

df['Grade'] = df['Average'].apply(grade) # apply grade function to each row in 'Average' column
```

```python
# sort dataframe by average marks - highest first
top_students = df.sort_values(by='Average',ascending=False)
top_students
# display sorted result
```

```python
plt.figure(figsize=(10,5))
# create figure with size
plt.bar(df['Name'], df['Average'])
# create bar chart (name vs average)
plt.title('Student Average Performance')
# set chart title
plt.xlabel('Students')
# label x axis
plt.ylabel('Average Marks')
# label y axis
plt.xticks(rotation=45)
# rotate names for readability
plt.show
# display the chart
```
